```
def a_star(start, goal, graph, h):

    open_list = []          # nodes to be checked
    closed_list = []        # nodes already checked

    open_list.append(start)

    g = {}                  # cost from start
    g[start] = 0

    parent = {}             # to store path
    parent[start] = None

    while open_list:

        # find node with smallest f = g + h
        current = open_list[0]
        for node in open_list:
            if g[node] + h[node] < g[current] + h[current]:
                current = node

        # if goal reached
        if current == goal:
            path = []
            while current is not None:
                path.append(current)
                current = parent[current]
            return path[::-1]

        open_list.remove(current)
        closed_list.append(current)

        # check neighbors
        for neighbor, cost in graph[current]:

            if neighbor in closed_list:
                continue

            new_g = g[current] + cost

            if neighbor not in open_list:
                open_list.append(neighbor)
                g[neighbor] = new_g
                parent[neighbor] = current

            elif new_g < g[neighbor]:
                g[neighbor] = new_g
                parent[neighbor] = current

    return None
graph = {
    'A': [('B', 5), ('C', 3)],
    'B': [('D', 1)],
    'C': [('D', 1)],
    'D': []
}

h = {
    'A': 3,
    'B': 2,
    'C': 1,
    'D': 0
}

print(a_star('A', 'D', graph, h))

```

```
# Connected graph (node -> neighbors)
graph = {
    'A': ['B', 'C'],
    'B': ['D'],
    'C': ['E'],
    'D': ['F', 'G'],
    'E': ['H', 'I'],
    'F': [],
    'G': [],
    'H': [],
    'I': []
}

# Terminal node values
values = {
    'F': 3,
    'G': 5,
    'H': 2,
    'I': 9
}

def minimax(node, is_max, visited):
    if node in values:
        return values[node]

    if node in visited:
        return visited[node]

    if is_max:
        best = -9999
        for n in graph[node]:
            best = max(best, minimax(n, False, visited))
    else:
        best = 9999
        for n in graph[node]:
            best = min(best, minimax(n, True, visited))

    visited[node] = best
    return best

visited = {}
result = minimax('A', True, visited)
print("Optimal value:", result)


```
